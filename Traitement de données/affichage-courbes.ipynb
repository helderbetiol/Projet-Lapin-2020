### Affichage de la courbe

```python
# Temps	PressionArterielle	PAmoyenne Spirometrie	FrequenceCardiaque	FrequenceRespiratoire
```

```python
valeurEtudiee = 'FrequenceRespiratoire' #rentrer le nom de la colonne à étudier
pas = 0.1 # Pas désiré en seconde. Pas minimum : 0.001 ; attention plus le pas est petit plus ça demande de la puissance. Pas besoin d'un pas tr_ès petit pour juste voir les tendances
```

```python
import matplotlib.pyplot as plt 

#On va étudier seulement un échantillon sinon trop long. Mais sur du long terme peut d'influence
nb_donnees = int(data['Temps'].size/(pas/0.001))
indices = [i*(pas/0.001) for i in range(nb_donnees)]
x = []
y = []
for i in indices :
    x.append(data['Temps'][i])
    y.append(data[valeurEtudiee][i])

# Création de la courbe
plt.plot(x,y)
plt.xlabel('Temps en secondes')
plt.ylabel(valeurEtudiee)

injection = data['Remarque'][0]
plt.title(injection)
plt.show()
```
